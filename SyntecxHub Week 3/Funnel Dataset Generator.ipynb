```python
## initialize Faker and empty list for record
fake = Faker()
data = []

```

```python
## Funnel Stages and probability
stages = ['Visit', 'Product View', 'Add To Cart', 'Purchase']
probabilities = {
    'Visit':1.0,
    'Product View':0.7,
    'Add To Cart':.35,
    'Purchase':0.6
}
```

```python
## Supporting data from randomize
devices = ['Mobile', 'Desktop', 'Tablet']
regions = ['South', 'Norht', 'East', 'West']
channels = ['Google Ads', 'Organic', 'Email', 'Social Media', 'Friends']
categories = ['Electronics', 'Fashion', 'Home', 'Beauty', 'Sport']
payment_methods = ['Card', 'Online', 'Credit Card']
```

```python
## Generate data for 10,000 persons
for i in range(1, 1001):
    user_id = f"USR{i:05d}"
    session_id = f"SES{i:05d}"
    event_time = fake.date_time_between(start_date = '-31d', end_date = 'now')

    reached_purchase = False
```

```python
for i in range(1, 10001):
    user_id = f"USR{i:05d}"
    session_id = f"SES{i:05d}"
    event_time = fake.date_time_between(start_date='-31d', end_date='now')

    reached_purchase = False

    # Simulate user progression through funnel stages
    for stage in stages:

        if random.random() < probabilities[stage]:

            record = {
                'User_ID': user_id,
                'Session_ID': session_id,
                'Activities': stage,
                'Timestamp': event_time.strftime('%Y-%m-%d %H:%M:%S'),
                'Devices': random.choice(devices),
                'Region': random.choice(regions),
                'Channel': random.choice(channels),
                'Product_Category': random.choice(categories),
                'Payment_Method': random.choice(payment_methods),
                'Revenue': round(random.uniform(200, 2000), 2) if stage == 'Purchase' else 0,
                'Bounce_Flag': 'No' if stage == 'Purchase' else 'Yes'
            }

            data.append(record)

            event_time += timedelta(
                minutes=random.randint(2, 7)
            )

            if stage == 'Purchase':
                reached_purchase = True

        else:
            break
```

```python
# Create DataFrame
df = pd.DataFrame(data)

# Identify users who completed a purchase
purchasing_users = df[df['Activities'] == 'Purchase']['User_ID']

# Update bounce flag
df['Bounce_Flag'] = df['User_ID'].isin(purchasing_users).map({
    True: 'No',
    False: 'Yes'
})

# Save dataset
df.to_csv("funnel_analysis_dataset.csv", index=False)

print("Funnel Dataset Generated Successfully! 💃💃💃")
```
